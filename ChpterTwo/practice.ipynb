```python
import pandas as pd

data = pd.read_csv("Salary_Data.csv")
data.head()
```

```python
data.describe()
data.info()
```

```python
# for btter understanding we can visualize the dta using a histogram
import matplotlib.pyplot as plt
data.hist(bins = 50, figsize=(12,8))
plt.show()
```

```python
# so the next thing is to create a test and train split, from the histogram I noticed that the dataset is a bit evenly distributed, not a lot of outliers and no abnormality.

from sklearn.model_selection import train_test_split
train_set, test_set =  train_test_split(data, test_size = 0.2, random_state = 42)
```

```python
# DATA VISUALIZATION
# we also have to create a copy of the dtaset before visulaizing because we dont want our experimentation to affect the dataset
from matplotlib.pyplot import scatter


data = train_set.copy()

data.plot(kind = "scatter", x="YearsExperience", y="Salary", grid =True,)
plt.show()
```

```python
#from the plot above we have a linear relationship between years of experience and salary
# the next thing is o look for correlation matrix
corr_matrix = data.corr()
corr_matrix["Salary"].sort_values(ascending=False)
```
